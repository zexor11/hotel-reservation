```python
# Select only numeric columns for correlation matrix
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(17, 15))
sns.heatmap(numeric_columns.corr(), annot=True)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x="booking_status", y="arrival_year", data=df)
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x="booking_status", y="repeated_guest", data=df)
```

```python
fig = plt.figure(figsize=(15, 5))

plt.subplot(231)
plt.title("subplot:231")
sns.countplot(data=df, x="no_of_weekend_nights", hue="booking_status")

plt.subplot(232)
plt.title("subplot:232")
sns.scatterplot(data=df, x="no_of_weekend_nights", y="no_of_week_nights", hue="booking_status")

plt.subplot(233)
plt.title("subplot:233")
sns.boxplot(data=df, x="booking_status", y="no_of_weekend_nights", hue="booking_status")
plt.xticks(rotation=45)  # Rotating x-axis labels for better visibility

plt.tight_layout()  # Adjust subplot layout for better spacing
plt.show()

```
